```python
def adj_rsquare(x,y):
    r2=model.score(x,y)
    n=x.shape[0]
    p=x.shape[1]
    
    adjusted_r_square=1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r_square


value=adj_rsquare(x_train,y_train)
value
```

```python
def linear_regression(data):
    
    x=data[["Avg. Area Income"]]
    y=data[["Price"]]
           
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,random_state=42)
    Reg_Model=LinearRegression()
    model=Reg_Model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    r_sqaure=r2_score(y_test,y_pred)
    
    MAE=mean_absolute_error(y_test,y_pred)
    MAPE=mean_absolute_percentage_error(y_test,y_pred)
    Accuracy=(1-mean_absolute_percentage_error(y_test,y_pred))*100
    MSE=mean_squared_error(y_test,y_pred)
    rmse=np.sqrt(mean_squared_error(y_test,y_pred))
           
    
    n=x.shape[0]
    p=x.shape[1]
    
    Adj_r_square=1-(1-r_sqaure)*(n-1)/(n-p-1)      
           
    list_1=[MAE,MAPE,MSE,Adj_r_square,Accuracy,rmse] 
    resultant_df=pd.DataFrame(list_1,index=["MAE","MAPE","MSE","Adj_r_square","Accuracy","rmse"],columns=["Performance_Metrics_Values"])       
    return resultant_df
           
linear_regression(data)
```
